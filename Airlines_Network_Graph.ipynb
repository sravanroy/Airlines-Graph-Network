# *Goal* :

* Implementing graph theory to solve an airline problem
* Find the shortest path between two airports in US by Distance and AirTime

# *Methodologies* :

*  The airlines data can be found [here](https://drive.google.com/file/d/1S5c2z89SdzxSVeflWuwVT9EWHXB7Iv8Y/view) 
*  Used dijkstra algorithm to find the shortest path between two nodes in a graph

# *Code and Implementation* :

### Loading required libraries and loading the data 

```python
import pandas as pd
import numpy as np

airport_data = pd.read_csv("data.csv")
```

```python
airport_data.head()
```

> CRSDepTime and CRSArrTime are scheduled departure and scheduled arrival times of flights

> DepTime and ArrTime are actual departure and actual arrival times of flights

### Loading NetworkX package to form graphs for the airport locations

```python
import networkx as nx
```

```python
df = nx.from_pandas_edgelist(airport_data,  source = 'Origin', target = 'Dest', edge_attr = True)
```

```python
df.nodes()
```

```python
df.edges()
```

### Plotting and visualizing the connections between the airports

```python
import matplotlib.pyplot as plt
%matplotlib inline


plt.figure(figsize=(12,12))
nx.draw_networkx(df, with_labels=True, font_size=16, node_size = 1000, linewidths = 0.5, font_color = 'white')
```

### Let's find the shortest path by distance from 'LAX' to 'PHL' airport using Dijkstra algorithm

```python
shortest_path_distance = nx.dijkstra_path(df, source = 'LAX', target = 'PHL', weight = "Distance")
shortest_path_distance
```

### Let's find the shortest path by airtime from 'LAX' to 'PHL' airport using Dijkstra algorithm


```python
shortest_path_airtime = nx.dijkstra_path(df, source = 'LAX', target = 'PHL', weight = "AirTime")
shortest_path_airtime
```
